`cleaning the data`

```python
#finding the null values
data.isnull().sum()
```

```python
#dropping the unwanted values
newdata=data.drop(['Country','Postal Code','Segment'], axis=1)
```

```python
newdata.head()
```

`Relationship analysis`

```python
#finding the correlation between different columns
correlation=newdata.corr()
```

```python
#visualization of correlation between different columns
sns.heatmap(correlation,xticklabels=correlation.columns, yticklabels=correlation.columns, annot=True)
```

```python
#plotting graphs between different columns
sns.pairplot(newdata)
```

```python
#relation plot between profit and sales
sns.relplot(y='Profit', x='Sales', hue='Category' ,data=newdata)
```

```python
#relation plot between sales and quantity according to ship mode
sns.relplot(x='Sales', y='Quantity', hue='Ship Mode' ,data=newdata)
```

```python
#relation between sales and quantity according to region
sns.relplot(x='Sales', y='Quantity', hue='Region' ,data=newdata)
```

```python
#relation between discount and profit according to sales
sns.relplot(x='Discount', y='Profit', hue='Sales' ,data=newdata)
```

```python
#relation between sales and profit according to region
sns.relplot(x='Sales', y='Profit', hue='Region' ,data=newdata)
```

```python
#displot of profit
sns.distplot(newdata['Profit'],bins=4)
```

```python
#displot of sales
sns.distplot(newdata['Sales'], bins=5)
```

```python
#boxplot of profit
sns.catplot(x='Profit', kind='box', data=newdata)
```

```python
#boxplot of sales
sns.catplot(x='Sales', kind='box', data=newdata)
```
